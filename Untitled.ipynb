```python
import pandas as pd
import os
```

```python
df = pd.read_csv("E:\\tensorflow\\simply neural\\Churn.csv")
```

```python
df.head()
```

```python
x = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
import numpy as np
```

```python
x =  np.asarray(x).astype(np.float32)
y =  np.asarray(y).astype(np.float32)
```

```python

y
```

```python
from sklearn.model_selection import train_test_split
```

```python
!pip install scikit-learn
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dropout, Dense
```

```python
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
x_train.shape
```

```python
np.expand_dims(y_train, axis=0).shape
```

```python
y_train.shape
```

```python
len(x_train)
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=(6575)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_train.shape
```

```python
model.summary()
```

```python
model.compile(loss='BinaryCrossentropy', optimizer='SGD', metrics='accuracy')
```

```python
model.fit(x_train, y_train, epochs=1, batch_size=32)
```

```python
y_pred = model.predict(x_test)
```

```python
ypred = [0 if val<0.5 else 1 for val in y_pred] 
```
